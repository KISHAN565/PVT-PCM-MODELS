```python
import numpy as np

def bairstow(poly, r, s, tol=1.0e-5, max_iter=100):

    n = len(poly) - 1
    roots = []

    while n > 2:
        # Initialize u and v
        u, v = np.zeros(n + 1), np.zeros(n + 1)
        u[-1], u[-2] = poly[-1], poly[-2] + r * poly[-1]

        # Iterative loop
        for _ in range(max_iter):
            for i in range(n - 2, -1, -1):
                u[i] = poly[i] + r * u[i + 1] + s * u[i + 2]

            v[-1], v[-2] = u[-1], u[-2] + r * u[-1]

            for i in range(n - 3, -1, -1):
                v[i] = u[i + 1] + r * v[i + 1] + s * v[i + 2]

            # Compute the deltas
            determinant = v[1] * v[1] - v[0] * v[2]
            if determinant == 0:
                raise ValueError("Determinant is zero. The method cannot proceed.")

            delta_r = (-u[0] * v[1] + u[1] * v[0]) / determinant
            delta_s = (-u[0] * v[2] + u[2] * v[0]) / determinant

            r += delta_r
            s += delta_s

            if abs(delta_r) < tol and abs(delta_s) < tol:
                break

        # Get the roots from the quadratic factor
        discriminant = r * r + 4 * s
        if discriminant >= 0:
            root1 = (-r + np.sqrt(discriminant)) / 2
            root2 = (-r - np.sqrt(discriminant)) / 2
            roots.extend([root1, root2])
        else:
            real_part = -r / 2
            imag_part = np.sqrt(-discriminant) / 2
            roots.extend([complex(real_part, imag_part), complex(real_part, -imag_part)])

        # Deflate the polynomial
        poly = u[2:]
        n -= 2

    # Handle the last quadratic or linear equation
    if n == 2:
        r = -poly[1] / poly[2]
        s = -poly[0] / poly[2]
        discriminant = r * r - 4 * s
        if discriminant >= 0:
            roots.append((-r + np.sqrt(discriminant)) / 2)
            roots.append((-r - np.sqrt(discriminant)) / 2)
        else:
            real_part = -r / 2
            imag_part = np.sqrt(-discriminant) / 2
            roots.extend([complex(real_part, imag_part), complex(real_part, -imag_part)])
    elif n == 1:
        roots.append(-poly[1] / poly[0])

    return roots


coefficients = [[0.7, -4, 6.2, -2], [-3.704,16.3,-21.97,9.34],[1,-2,6,-2,5]] # Coefficients of all polynomials
initial_r = 0.5 #intial r value guess
initial_s = -0.5 #intial s value guess
roots = []
for i in range(len(coefficients)):
    roots.append(bairstow(coefficients[i], initial_r, initial_s)) #appending roots in list

print("Roots of the polynomial 0.7x^3 − 4x^2 + 6.2x − 2 are:") #printing the list
for root in roots[0]:
    print(root)
print("Roots of the polynomial -3.704x^3 + 16.3x^2 − 21.97x + 9.34 are:")
for root in roots[1]:
    print(root)
print("Roots of the polynomial x^4 − 3x^3 − 7x^2 + 27x − 18 are:")
for root in roots[2]:
    print(root)

```

```python
import numpy as np
import pandas as pd
# Function to calculate Lagrange basis polynomial
def lagrange_basis(x, i, x_values):
    n = len(x_values)
    L = 1.0
    for j in range(n):
        if j != i:
            L *= (x - x_values[j]) / (x_values[i] - x_values[j])
    return L

# Function to calculate interpolated value at a given point
def lagrange_interpolation(x_values, y_values, x):
    n = len(x_values)
    y = 0.0
    for i in range(n):
        L = lagrange_basis(x, i, x_values)
        y += L * y_values[i]
    return y
file = '/soil consolidation data.xlsx - Table.csv'
data = pd.read_csv(file)
data.head()
```

```python
import numpy as np

# Function to calculate the divided difference table
def divided_diff_table(x_values, y_values):
    n = len(y_values)
    # Create a table initialized with zeros
    diff_table = np.zeros((n, n))
    # Fill the first column with y_values
    diff_table[:, 0] = y_values

    # Fill the rest of the table
    for j in range(1, n):
        for i in range(n - j):
            diff_table[i][j] = (diff_table[i+1][j-1] - diff_table[i][j-1]) / (x_values[i+j] - x_values[i])

    return diff_table

# Function to calculate the interpolated value using Newton's Divided Difference method
def newton_interpolation(x_values, y_values, x_unknown):
    # Get the divided difference table
    diff_table = divided_diff_table(x_values, y_values)

    # First value of the polynomial is the first value in the divided difference table
    result = diff_table[0, 0]

    # Iteratively add each term to the polynomial
    n = len(x_values)
    term = 1.0
    for i in range(1, n):
        term *= (x_unknown - x_values[i-1])
        result += term * diff_table[0, i]

    return result

# Assigning usage with 11 known data points
x_values = data['Time']  # Known x values
y_values = data['Settlement']  # Corresponding y values
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
link ='/Stress-strain data.xlsx - Sheet1.csv'
table = pd.read_csv(link)
table.head()
# Hyperbolic model function
def hyperbolic_model(x, a, b):
    return (x / (a +( b * x)))

#stress-strain data
strain = table['Strain (%)']  # Strain (x)
stress = table['Stress(kPa)']  # Stress (y)

# Fit the data to the hyperbolic model
params, covariance = curve_fit(hyperbolic_model, strain, stress)

# Extract the parameters a and b
a, b = params
print(f"Fitted parameters: a = {a:.4f}, b = {b:.4f}")
# Generate fitted stress values for plotting
strain_fine_1 = np.linspace(min(strain), max(strain), 100)
stress_fitted_1 = hyperbolic_model(strain_fine, a, b)

# Plot the data and the fitted model
plt.figure(figsize=(8, 6))
plt.plot(strain, stress, 'o', label='Original Data', color='red', markersize=8)
plt.plot(strain_fine_1, stress_fitted_1,label='Fitted curve', color='blue')
plt.xlabel('Strain')
plt.ylabel('Stress(in kPa)')
plt.title('Hyperbolic Model Fit for Stress-Strain Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example stress-strain data
strain = table['Strain (%)']  # Strain (x)
stress = table['Stress(kPa)']  # Stress (y)

# Degree of the polynomial
degree = 9

# Create the Vandermonde matrix (X matrix)
X = np.vander(strain, degree + 1)

# Solve the normal equation (X^T * X) * c = X^T * y to get the coefficients
coefficients = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(stress)

# Print the polynomial coefficients
print(f"Polynomial coefficients: {coefficients}")

# Predict the stress values (fitted values) for plotting
strain_fine = np.linspace(min(strain), max(strain), 100)
X_fine = np.vander(strain_fine, degree + 1)
stress_fitted = X_fine.dot(coefficients)

# Plot the original data and the fitted polynomial curve
plt.figure(figsize=(8, 6))
plt.plot(strain, stress, 'o', label='Original Data', color='red', markersize=8)
plt.plot(strain_fine, stress_fitted, label=f'Fitted Polynomial of Degree {degree}', color='blue')
plt.xlabel('Strain')
plt.ylabel('Stress in (kPa)')
plt.title(f'Polynomial Regression (Degree {degree}) Fit for Stress-Strain Data')
plt.legend()
plt.grid(True)
plt.show()
```
